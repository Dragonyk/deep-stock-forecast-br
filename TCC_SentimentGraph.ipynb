```python
gn = '/content/drive/MyDrive/DEV/Python/Datasets/TCC/NEWTESTS/PETR4/sentiment_gn_PETR4_2008-01-01_2022-09-20.csv'
tw = '/content/drive/MyDrive/DEV/Python/Datasets/TCC/NEWTESTS/PETR4/sentiment_tw_PETR4_2008-01-01_2022-09-20.csv'

df_gn = pd.read_csv(gn)
df_tw = pd.read_csv(tw)
```

```python
stest = 'temp.csv'
df_test = pd.read_csv(stest)

test_gn_sentiments = []
test_tw_sentiments = []

def check_sentiment(negative, positive):
    if negative > positive:
        return "negative"
    elif positive > negative:
        return "positive"
    else:
        return "neutral"

row_size = len(df_test.index)
for i in range(0, row_size):
  gn_neg = df_test.loc[i, 'negative_gn']
  gn_pos = df_test.loc[i, 'positive_gn']
  tw_neg = df_test.loc[i, 'negative_tw']
  tw_pos = df_test.loc[i, 'positive_tw']
  test_gn_sentiments.append(check_sentiment(gn_neg,gn_pos))
  test_tw_sentiments.append(check_sentiment(tw_neg,tw_pos))

df_sentiment = pd.DataFrame(list(zip(test_gn_sentiments, test_tw_sentiments)),
              columns =['gn_sentiment', 'tw_sentiment'])
```

```python
df_sentiment.head()
```

```python
def count_values_in_column(data,feature):
    total=data.loc[:,feature].value_counts(dropna=False)
    percentage=round(data.loc[:,feature].value_counts(dropna=False,normalize=True)*100,2)
    return pd.concat([total,percentage],axis=1,keys=['Total','Percentage'])
```

```python
pichart = count_values_in_column(df_sentiment,"gn_sentiment")
names= pichart.index
size=pichart["Percentage"]
 
plt.pie(size, labels=names, colors=['green','red','skyblue'], autopct="%1.1f%%", shadow=True)
p=plt.gcf()
plt.show()

pichart = count_values_in_column(df_sentiment,"tw_sentiment")
names= pichart.index
size=pichart["Percentage"]
 
plt.pie(size, labels=names, colors=['green','red','skyblue'], autopct="%1.1f%%", shadow=True)
p=plt.gcf()
plt.show()
```
