```python
import pandas as pd
```

```python
curstock = 1

filename = 'VALE3_gn_mergeall'

merge_files = ['raw_gn_VALE3_2008-01-01_2022-09-20 (4).csv',
               'raw_gn_VALE3_2008-01-01_2022-09-20 (3).csv',
               'raw_gn_VALE3_2008-01-01_2022-09-20 (2).csv',
               'raw_gn_VALE3_2008-01-01_2022-09-20 (1).csv']



df_files = []
for item in merge_files:
  df_files.append(pd.read_csv(item))
# -----------

df_merge = pd.concat(df_files, ignore_index=True)
df_merge = df_merge.apply(lambda x: x.astype(str).str.lower()).drop_duplicates(subset=['text'], keep=False)

df_merge = df_merge[df_merge["text"].str.contains("cripto") == False]

df_merge = df_merge[df_merge["text"].str.contains("bitcoin") == False]

df_merge['date'] = pd.to_datetime(df_merge['date'])
df_merge = df_merge.sort_values(by='date')

df_merge.to_csv(filename+'.csv', sep=',', index=False)



print(df_merge.head())
```
